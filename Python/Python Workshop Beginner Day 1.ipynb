## Math & Logic
![Math](https://media1.giphy.com/media/v1.Y2lkPTc5MGI3NjExdnQzZTBxd3Vka2toNm9hYzRzOHMycjVqM2ltMGU1aTVkeDhvbXVjcCZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/APqEbxBsVlkWSuFpth/giphy.gif)

Python is a great calculator! You can do common math operations right away: ‚ûï‚ûñ‚úñÔ∏è‚ûó

- `+` ‚Üí addition  
- `-` ‚Üí subtraction  
- `*` ‚Üí multiplication  
- `/` ‚Üí division (always gives a float)  
- `//` ‚Üí floor division (whole number result)  
- `%` ‚Üí modulus (remainder after division)  
- `**` ‚Üí exponent (e.g., `2**3` means 2¬≥ = 8)

````python
a = 17
b = 5

print("Addition (a + b):", a + b)
print("Subtraction (a - b):", a - b)
print("Multiplication (a * b):", a * b)
print("Division (a / b):", a / b)          # ‚Üí 3.4
print("Floor Division (a // b):", a // b)   # ‚Üí 3
print("Modulus (a % b):", a % b)           # ‚Üí 2 (remainder)
print("Exponent (a ** 2):", a ** 2)        # ‚Üí 289
````

````python
# Comparisons return True or False
print("5 > 3 ?", 5 > 3)        # True
print("10 == 10 ?", 10 == 10)  # True
print("'a' != 'b' ?", 'a' != 'b')  # True

# These are used in if statements!
is_adult = age >= 18  # This is a boolean!
````

### Lists: Your First Collection

Lists hold **multiple items** in order ‚Äî like a shopping list! üì¶

Use **bracket `[]`**.

````python
colors = ["red", "green", "blue"]
print("Original list:", colors)

colors.append("purple")  # Add to end
print("After adding purple:", colors)

colors[0] = "orange"  # Change first item
print("After changing red ‚Üí orange:", colors)

print("Number of colors:", len(colors))
````

<a id="tuples"></a>
## Tuples: Lists That Don‚Äôt Change üõ°Ô∏è

Tuples are like **lists**, but **you can‚Äôt change them** after creating (they‚Äôre *immutable*). Great for coordinates, RGB values, or fixed data!

Use **parentheses `()`**.

````python
point = (10, 20)  # (x, y)
rgb = (255, 128, 0)  # orange!

print("Point:", point)
print("Red value:", rgb[0])

# Unpack tuple
x, y = point
print(f"x = {x}, y = {y}")

# Convert to list if you need to change it
mutable_point = list(point)
mutable_point[0] = 15
print("As list with new x:", mutable_point)
````

<a id="dictionaries"></a>
## Dictionaries: Key-Value Magic üîë‚û°Ô∏èüì¶

Store data as **key-value pairs**‚Äîlike a real dictionary!
- **Key**: unique label (`"name"`)
- **Value**: data (`"Sam"`)

Use **curly braces `{}`**.

````python
student = {
    "name": "Sam",
    "age": 22,
    "courses": ["Math", "Python"],
    "active": True
}

print("Student:", student)
print("Name:", student["name"])

# Add new key
student["email"] = "sam@example.com"
print("\nAfter adding email:", student)

# Safe get (won‚Äôt crash if key missing)
phone = student.get("phone", "Not provided")
print("Phone:", phone)

print("\nKeys:", list(student.keys()))
````

<a id="conditionals"></a>
## Make Decisions with `if`/`else`

Your code can **choose paths** based on conditions! ü§î

![Decision](https://tadabase-docs-assets.s3.amazonaws.com/uploads/images/gallery/2020-05/657-b782f3286409b190756c1b5dc8940285560896fa.gif)

````python
# Example without input (safe for all environments)
temperature = 25

if temperature > 30:
    print("It's hot! ‚òÄÔ∏è")
elif temperature > 20:
    print("Perfect weather! üòé")
else:
    print("Grab a jacket! üß•")

# You can change `temperature` above and re-run!
````

<a id="loops"></a>
## Loops: Do It Again!

Loops **repeat actions** ‚Äî great for counting, listing, or waiting! üîÅ

![Loop](https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExcm1vc2p5YmRlaGNrcTRsamZkdDZtbGo4NWRxODY5dXZqN2xhcXN0cCZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/6HsjDOBPwY1eIS6kE0/giphy.gif)

````python
# üîÅ for loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I love {fruit}s!")

print("\n---\n")

# ‚è≥ while loop
count = 3
while count > 0:
    print(f"Countdown: {count}")
    count -= 1  # decrease by 1
print("Blast off! üöÄ")

print("\n---\n")

# ‚ö†Ô∏è Be careful with while loops ‚Äî avoid infinite loops!
# Example: don‚Äôt forget to update the condition variable!
````

## Mini Project: Simple Calculator

Put it all together! This calculator adds, subtracts, multiplies, or divides two numbers. üßÆ

![Calculator](https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExa2hxODExdWJ6dGVueGV2ZW9tNWVycWxoeDFlbDZnejA1eHBkbWJwZSZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/EAAKR8I15yFg0RK5as/giphy.gif)

````python
def calculator():
    print("\n--- üßÆ Simple Calculator ---")
    try:
        num1 = float(input("Enter first number: "))
        op = input("Operator (+, -, *, /): ")
        num2 = float(input("Enter second number: "))
        
        if op == "+": result = num1 + num2
        elif op == "-": result = num1 - num2
        elif op == "*": result = num1 * num2
        elif op == "/":
            if num2 != 0:
                result = num1 / num2
            else:
                result = "Error: Can't divide by zero!"
        else:
            result = "Invalid operator!"
        
        print("Result:", result)
    except:
        print("Please enter valid numbers!")

# Uncomment to try (if input works in your environment):
# calculator()
````

## Final Challenge!

**Task**: Print all **even numbers from 1 to 20**. üèÜ

Try it yourself first‚Ä¶ then check the solution below!

![Challenge](https://media.giphy.com/media/v1.Y2lkPTc5MGI3NjExN3Q2Z3c5ZGhxYW00ZmE4dGwxcnJlNnphY3pka3FhemZyemlocTUwZyZlcD12MV9naWZzX3NlYXJjaCZjdD1n/88ItEmRSOqJnQXSlOe/giphy.gif)

````python
# Solution
print("Even numbers from 1 to 20:")
for n in range(1, 21):
    if n % 2 == 0:
        print(n)

# Fancy one-liner (bonus!)
print("\nBonus with list comprehension:", [n for n in range(1, 21) if n % 2 == 0])
````
